```python
x = rating['user-id'].value_counts()>200
```

```python
y = x[x].index
```

```python
y.shape
```

```python
ratings = rating[rating['user-id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
rating_with_books = ratings.merge(books,on='ISBN')
```

```python
rating_with_books.shape
```

```python
rating_with_books.head()
```

```python
number_rating =rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'number_of_ratings'},inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = rating_with_books.merge(number_rating,on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['number_of_ratings']>50]
```

```python
final_rating.shape
```

```python
final_rating.duplicated().sum()
```

```python
final_rating.drop_duplicates(['user-id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user-id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[432,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
def Recommend(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestions)):
            print(book_pivot.index[suggestions[i]])
```

```python
Recommend('Animal Farm')
```
